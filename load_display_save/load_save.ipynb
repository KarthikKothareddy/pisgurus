```python
def display(image, flip=False, cmap=None):
    if flip:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 10))
    _ = plt.imshow(image, cmap=cmap)
    plt.axis("off")
```

```python
display(image, True)
```

```python
# changing pixel at (0, 0)
image[0, 0] = (0, 0, 255)
display(image, True)
```
