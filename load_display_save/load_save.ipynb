```python
image = cv2.imread("sunflower.png")
```

```python
(h, w) = image.shape[:2]
```

```python
h, w
```

```python
(b, g, r) = image[0, 0]
b, g, r
```

```python
def display(image, flip=True, cmap=None):
    if flip:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 10))
    _ = plt.imshow(image, cmap=cmap)
    plt.axis("off")
```

```python
display(image, True)
```

```python
# changing pixel at (0, 0)
image[0, 0] = (0, 0, 255)
display(image, True)
```

```python
# access portions of an image through slicing

# get the center by getting half from both co-ordinates
(cX, cY) = (w//2, h//2)

# select top left half
display(image[0:cY, 0:cX])
```

```python
# top right
tr = image[0:cY, cX:w]
# display(tr)
# bottom left
bl = image[cY:h, 0:cX]
# display(bl)
# bottom right
br = image[cY:h, cX:w]
# display(br)
```

```python
# make topleft corner blue
tl = image[0:cY, 0:cX]
image[0:cY, 0:cX] = (255, 0, 0)
display(image)
```
